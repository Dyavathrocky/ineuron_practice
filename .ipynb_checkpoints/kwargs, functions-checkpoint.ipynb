```python
def genfib(n):
    a=1
    b=1
    output = []
    for i in range(n):
        output.append(b)
        a, b = b, a+b
    return output
```

```python
genfib(23)
```

```python
def genfib(n):
    a=1
    b=1
    for i in range(n):
        yield a
        a, b = b, a+b
```

```python
genfib(6)
```

```python
for i in genfib(6000):
    print(i)
```
